### <b>  <font color=   #33691e  > <i> Operador lógico "and"

| Condición 1 | Condición 2 | and |
| ------ |---------| ------|

| False| True    | False   |

<b> Ejemplo de uso del operador lógico "and"

## <font color=  #0000FF   > <i> Funciones de Órden Superior

Son aquellas que contienen otras funciones como parámetros de entrada o si devuelve una función como salida, es decir, las funciones que operan con otra función.

Funciones principales son **filter(), map() y reduce()**.

###  <font color=#33691e > <i>  Función Filter

La función **filter()** devuelve un iterador donde los elementos se filtran a través de una función para probar si el elemento es aceptado o no.

```python
#Función para otener números impares

numeros = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

def es_impar(num):
    return num%2 !=0

list(filter(es_impar, numeros))
```

```python
#Función para otener números pares

def es_par(num):
    return num%2 == 0

list(filter(es_par, numeros))
```

###  <font color=#33691e > <i> Función lambda

Las funciones lambda se conocen también como funciones "anónimas", "on the go", "on demand" u "online"

No es necesario asignarles un nombre

Están formadas por una sola línea

Suelen utilizarse para codificar funciones muy sencillas, no es posible codificar funciones complejas, por ejemplo, aquellas que contienen bucles o sentencias if-else


```python
#Función común
def area_triangulo(base, altura):
    return (base*altura)/2

print(f'El área del triángulo es {area_triangulo(4,5)}')
```

Su estructura es la siguiente:

```python
#Función lambda

area_triangulo2 = lambda base, altura: (base*altura)/2

print(f'El área del triángulo es {area_triangulo2(4,5)}')

```

###  <font color=#33691e > <i>  Combinando las funciones Filter y Lambda

```python
numeros = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

#Mostramos los números impares
list(filter(lambda num: num % 2 != 0, numeros))
```

```python
#Mostramos los números pares
list(filter(lambda num: num % 2 == 0, numeros))
```

<br>

###  <font color=#33691e > <i> Función map

La función **map aplica una función a cada elemento** de una lista o tupla iterable

```python
numeros = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]

def duplicar(num):
    return num*2

list(map(duplicar, numeros))
```

###  <font color=#33691e > <i>  Combinando las funciones Map y Lambda

```python
list(map(lambda num: num*2, numeros))
```

## <font color= #0000FF  > <i> Variables Globales y Variables Locales

###  <font color=#33691e > <i>  Variables Locales

<b> Son aquellas que:

    - Se crean dentro de una función o le llegan a esta como parámetros.
    - Existen solamente mientras la función se ejecuta.

Ejemplo 1: Creamos una variable dentro de la función y tratamos de imprimirla fuera de ella

```python
def sumando(a, b):
    y = a+b
    print(y)

sumando(3,4)

print(y)
```

Ejemplo 2: Creamos una variable dentro de una función, la "sacamos" de la función con un return y tratamos de imprimirla

```python
def sumando(a, b):
    y = a+b
    print(y)
    return(y)

sumando(3,4)

print(y)
```

Ejemplo 3: Creamos una variable dentro de una función, la "sacamos" de la función con un return, la almacenamos en la variable z y la imprimimos en pantalla

```python
def sumando(a, b):
    y = a+b
    print(y)
    return(y)

z = sumando(3,4)

print(z)
```

###  <font color= #33691e > <i>  Variables Globales

<b> Son aquellas que:

    - Se crean afuera de una función.
    - Existen desde que son declaradas hasta que se termina el programa.

Ejemplo 1: Creamos las variables c y d afuera de la función y dentro de la función las sumamos

```python
# Dreclaración de variables globles fuera de la función
c=8
d=2

def sumando():
    y = c+d
    print(f'y = {y}')
    return(y)

z = sumando()

print(f'z = {z}')
```

```python
print(f'c = {c}')
```

```python
print(f'd = {d}')
```

Ejemplo 1: Creamos las variables c y d afuera de la función, dentro de la función alteramos su valor y las sumamos

```python
# Variables globales declaradas al inicio
c=8
d=2

def funcion_1():
    c=3
    print(f'c dentro de la funcion_1 = {c}')

def funcion_2():
    d=5
    print(f'd dentro de la funcion_2 = {d}')

funcion_1()
funcion_2()
```

```python
print(f'c fuera de la funcion = {c}')
```

```python
print(f'd fuera de la funcion = {d}')
```

#### <font color= #990000 > <i>  Modificando una Variable Global desde una función

```python

c=8
d=2

def sumando():
    global c # Declara c como global
    c=3      # Modifica la variable global dentro de la función
    y = c+d
    print(y)
    return(y)


z = sumando()

print(z)
print(c)
```

```python
print(c)
```

```python
print(d)
```

## <font color= #0000FF  > <i> Diccionarios

###  <font color= #33691e > <i>  Definición y elementos que los forman

 Los diccionarios son **otra estructura de datos**, parecidas a las listas y a las tuplas, cuya principal característica es que los datos se almacenan asociados a campos o claves, de manera que se crea una asociación <font color=  #a93226  > <i>clave:valor</i> </font> para cada elemento.

Los elementos de un diccionario pueden estar en cualquier órden.

Permiten almacenar valores de diferentes tipos, e incluso listas y otros diccionarios dentro de ellos.
